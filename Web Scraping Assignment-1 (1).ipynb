#### 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
#requesting and parsing the content of the page
page= requests.get('https://www.imdb.com/chart/top/')
content= BeautifulSoup(page.content)

#creating the list
title=[]
rating=[]
year=[]

#data of top 100 movies names
for movie_name in content.find_all('td',class_='titleColumn'):
    c= movie_name.find('a')
    title.append(c.text)
    
#data of top 100 movies released year
for movie_year in content.find_all('td',class_='titleColumn'):
    c= movie_year.find('span')
    year.append(c.text.replace('(','').replace(')',''))
    
#data of top 100  movies rating   
for movie_rating in content.find_all('td',class_='ratingColumn imdbRating'):
    rating.append(movie_rating.text.replace("\n",''))
    
#creating Dataframe
top_rated_movies= pd.DataFrame({"Title": title,'Rating':rating,'Year':year})
top_rated_100_movies=top_rated_movies[:100]
top_rated_100_movies
```

#### 4) Write a python program to scrape product name, price and discounts from https://meesho.com/bagsladies/pl/p7vbp . 

```python
#requesting and parsing the content of the page
page= requests.get('https://meesho.com/bags-ladies/pl/p7vbp')
content= BeautifulSoup(page.content)

#creating list for collecting the required data
product=[]
price=[]
discounts=[]

#collecting the data (names, rate, discounts)
for i in content.find_all('div',class_='sc-dkPtRN ProductList__GridCol-sc-8lnc8o-0 FjWWx jMkQHN'):
    names= i.find('p')
    rate= i.find('h5',class_='Text__StyledText-sc-oo0kvp-0 hiHdyy')
    discount=i.find('p',class_='Text__StyledText-sc-oo0kvp-0 fCJVtz NewProductCard__DiscountTextParagraph-sc-j0e7tu-16 kmYsnm NewProductCard__DiscountTextParagraph-sc-j0e7tu-16 kmYsnm')
    product.append(names.text)
    price.append(rate.text.replace('₹',''))
    discount=discount.text.split(' ')
    discounts.append(discount[0].replace('₹',''))
    
#creating Dataframe
Handbag_price= pd.DataFrame({"Product": product,'Discount on first order':discounts,'Price':price})
Handbag_price
```

#### Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:

```python
def ODI_Team(gender):
    url_team=f'https://www.icc-cricket.com/rankings/{gender}/team-rankings/odi'
    url_players=f'https://www.icc-cricket.com/rankings/{gender}/player-rankings/odi'
    
    #requesting and parsing the content for top 10 ODI teams
    page= requests.get(url_team)
    content= BeautifulSoup(page.content)

    #creating the list of details required
    matches=[]
    points=[]
    rating=[]
    position=[]


    for table in content.find_all('table',class_='table'):
    
        #scrapping the top odi teams name
        teams=[team.text for team in table.find_all('span',class_='u-hide-phablet')] 
        #scrapping top 1 teams data
        matches.append(table.find('td',class_='rankings-block__banner--matches').text.replace('\n',''))
        points.append(table.find('td',class_='rankings-block__banner--points').text.replace('\n',''))
        rating.append(table.find('td',class_='rankings-block__banner--rating u-text-right').text.replace('\n',''))
    
        #collecting the remaining top teams data
        tables=[]
        for match in table.find_all('td',class_='table-body__cell u-center-text'):
            tables.append(match.text)
        for num in range(len(tables)):
            if (num%2)==0:
                matches.append(tables[num])
            else:
                points.append(tables[num])
        for rate in table.find_all('td',class_='table-body__cell u-text-right rating'):
            rating.append(rate.text) 
    
    position= np.arange(1,len(teams)+1) 

    #creating dataframe for the collected ODI team data
    top_ODI_team= pd.DataFrame({'Position':position,'Teams':teams,'Matches':matches,'Points':points,'Rating':rating})
    top_ODI_team.set_index('Position',inplace=True)
    top_10_ODI_team= top_ODI_team[:10]
    
    
    #requesting and parsing the content of the top odi batsmen and bowlers
    page= requests.get(url_players)
    content= BeautifulSoup(page.content)

    #seperating the html of both batting and bowling tables since it has same attributes
    for sector in content.find_all('h4'):
        if sector.text =='ODI Batting Rankings':
            batting_html= sector.find_parent('div').find_parent('div').find_parent('div')
        elif sector.text=='ODI Bowling Rankings':
            bowling_html= sector.find_parent('div').find_parent('div').find_parent('div')
        
    #scrapping top odi batsmen data(name,team,rating)        
    batting_players= [bat_name.text.replace('\n','') for bat_name in batting_html.find_all('td',class_='table-body__cell name')]
    batting_players_team=[bat_team.text.replace('\n','') for bat_team in batting_html.find_all('span',class_='table-body__logo-text')]
    batting_players_rating=[bat_rat.text.replace('\n','') for bat_rat in batting_html.find_all('td',class_='table-body__cell u-text-right rating')]

    #scrapping top odi bowlers data(name,team,rating)        
    bowling_players=[bow_name.text.replace('\n','') for bow_name in bowling_html.find_all('td',class_='table-body__cell name')]
    bowling_players_team=[bow_team.text.replace('\n','') for bow_team in bowling_html.find_all('span',class_='table-body__logo-text')]
    bowling_players_rating=[bow_rat.text.replace('\n','') for bow_rat in bowling_html.find_all('td',class_='table-body__cell u-text-right rating')]

    #scrapping of number 1 batmen and bowlers data
    batting_players.append(batting_html.find('div',class_='rankings-block__banner--name').text)
    bowling_players.append(bowling_html.find('div',class_='rankings-block__banner--name').text)
    bat_top=batting_html.find('div',class_='rankings-block__banner--nationality').text.replace('\n','').split()
    bow_top=bowling_html.find('div',class_='rankings-block__banner--nationality').text.replace('\n','').split()
    batting_players_team.append(bat_top[0])
    batting_players_rating.append(bat_top[1])
    bowling_players_team.append(bow_top[0])
    bowling_players_rating.append(bow_top[1])

    position= np.arange(1,len(batting_players)+1)

    #creating dataframe for top 10 odi batsmen and bowlers
    top_10_ODI_batsmen= pd.DataFrame({'Players':batting_players,'Team':batting_players_team,'Rating':batting_players_rating})
    top_10_ODI_bowlers= pd.DataFrame({'Players':bowling_players,'Team':bowling_players_team,'Rating':bowling_players_rating})

    top_10_ODI_batsmen.sort_values('Rating',ascending=False,inplace=True)
    top_10_ODI_bowlers.sort_values('Rating',ascending=False,inplace=True)

    top_10_ODI_batsmen.set_index(position,inplace=True)
    top_10_ODI_bowlers.set_index(position,inplace=True)
    
    #returning the top 10 odi teams,batmens,bowlers data
    return top_10_ODI_team,top_10_ODI_batsmen,top_10_ODI_bowlers
```

#### 7) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post

```python
#requesting and parsing the content of the page
page= requests.get('https://coreyms.com/')
content= BeautifulSoup(page.content)


#scrapping the heading, date,content from each article 
heading= [title.text for title in content.find_all('h2',class_='entry-title')]
published_date= [date.text for date in content.find_all('time',class_='entry-time')]
article_content= [par.text.replace('\n','') for par in content.find_all('div',class_='entry-content')]

#some article doesn't have video link thus filling with none
video_link=[]
for vd in content.find_all('div',class_='entry-content'):
    if vd.find('span',class_='embed-youtube')==None:
        video_link.append(None)
    else:
        video_link.append(vd.find('iframe',class_='youtube-player').get('src'))

#creating dataframe for the above collected data
coreyms= pd.DataFrame({'Heading':heading,'Published Date':published_date,'Article Content':article_content,'Video Link':video_link})
coreyms= HTML(coreyms.to_html(render_links=True,escape=False))
coreyms
```

#### 9) Write a python program to scrape mentioned details from dineout.co.in :
#### i) Restaurant name
#### ii) Cuisine
#### iii) Location
#### iv) Ratings
#### v) Image URL 

```python
#requesting and parsing the content of the page
page= requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
content= BeautifulSoup(page.content)

#colleting the restaurants data
resta=[name.text for name in content.find_all('a',class_='restnt-name ellipsis')]
cuisine= [detail.text.split('|')[1] for detail in content.find_all('span',class_='double-line-ellipsis')]
location=[loc.text for loc in content.find_all('div',class_='restnt-loc ellipsis')]
rating= [ rate.text for rate in content.find_all('div',class_='restnt-rating rating-4')]
image_url= [ image.get('data-src') for image in content.find_all('img',class_='no-img')]

#creating the dataframe 
Restaurants= pd.DataFrame({'Name':resta,'Cuisine':cuisine,'Location':location,'Rating':rating,'Image URL':image_url})
Restaurants=HTML(Restaurants.to_html(render_links=True, escape=False))
Restaurants
```
